Information Retrieval

- Information Retrieval (IR) is finding material (usually documents) of an unstructured nature (usually text) that satisfies an information need from within large collections (usually stored on computers).

- Information need: a statement of what a user wants to know
- Document: a piece of text (or other medium) that is stored in a computer
- Collection: the total set of documents to be searched
- Unstructured nature: the text of the document has no internal structure that can be used to restrict the search


Simply put, the task is finding relevant documents/information in a collection of documents.

# What are Large Language Models?

Let's ask ChatGPT

Slide 1: Introduction to Large Language Models

- AI systems for natural language processing
- Based on deep learning techniques
- Massive number of parameters (billions)

Slide 2: Learning and Training Methods

- Unsupervised learning
- Maximum likelihood estimation (MLE)
- Masked language modeling

Slide 3: Architectures and Examples

- Transformer-based models
- OpenAI GPT series (GPT-3, GPT-4)
- Google BERT

Slide 4: Key Features and Applications

- Grammar and semantic learning
- Factual knowledge extraction
- Text generation, translation, summarization, question-answering, sentiment analysis

Slide 5: Attention Mechanisms

- Selective focus on relevant input data
- Improves performance on complex tasks

Slide 6: Challenges and Concerns

- High computational requirements
- Energy consumption
- Accessibility for researchers with limited resources
